```python
import pandas as pd
title_basics = pd.read_csv(r'C:\Users\DELL\Documents\GitHub\Business-Problem\Data\filtered_movies_basics.csv')
title_basics.info()
title_basics.head()
```

```python
## create a col with a list of genres
title_basics['genres_split'] = title_basics['genres'].str.split(',')
title_basics
```

```python
exploded_genres = title_basics.explode('genres_split')
exploded_genres

```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres

```

```python
title_genres = exploded_genres[['tconst','genres_split']]
title_genres
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map

```

```python
## make new integer genre_id and drop string genres
pd.options.mode.copy_on_write = True 
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns='genres_split')

```

```python
title_genres
```

```python
genre_lookup = pd.DataFrame({'genre_name':list(genre_map.keys()),'genre_id':list(genre_map.values())})
genre_lookup
```

```python
## Example
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = df['imdb_id'].fillna('').map(len).max()
cert_len = df['certification'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "imdb_id": CHAR(key_len+1), 
    "budget": FLOAT(),
    "revenue": FLOAT(),
    'certification':CHAR(key_len+1)}

```

```python
# Assuming `tmdb_data` is your DataFrame
columns_to_keep = ['imdb_id', 'revenue', 'budget', 'certification']

# Select only the needed columns
df = df[columns_to_keep]
```

```python
# Save to sql with dtype and index=False
df.to_sql('tmdb_data',conn, dtype=df_schema, if_exists='replace',index=False)

```
