```python
import pandas as pd
basics = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep='\t', low_memory=False)

```

```python
basics.head()
```

```python
akas = pd.read_csv('title-akas-us-only.csv',low_memory=False)
akas.head()
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
filter_us_titles = basics['tconst'].isin(akas['titleId'])
basics = basics[filter_us_titles]
basics
```

```python
import numpy as np
basics = basics.replace({'\\N':np.nan})
basics
```

```python
basics = basics.dropna(subset=['runtimeMinutes', 'genres'])
basics
```

```python
filtered_basics = basics[basics['titleType'] == 'movie']
filtered_basics
```

```python
basics = filtered_basics
basics
```

```python
basics['startYear'].value_counts()
```

```python
basics.loc[:-1, 'startYear'] = basics['startYear'].astype(float)
```

```python
basics['startYear'].value_counts()
```

```python
filtered_df = basics[(basics['startYear'] >= 2000) & (basics['startYear'] <= 2022)]
filtered_df 
```

```python
basics = filtered_df
basics
```

```python
filter_documentaries = basics['genres'].str.contains('Documentary')
# Exclude movies in the documentary category.
basics = basics[~filter_documentaries]
basics
```

```python
basics.info()
```

```python
basics.head()
```

```python
basics.to_csv("Data/filtered_movies_basics.csv", index=False)
```

```python
rating = pd.read_csv('https://datasets.imdbws.com/title.ratings.tsv.gz', sep='\t', low_memory=False)
rating
```

```python
filter_basics = rating['tconst'].isin(basics['tconst'])
rating = rating[filter_basics]
rating


```

```python
rating = rating.replace({'\\N':np.nan})

```

```python
rating
```

```python
rating.to_csv("Data/filtered_rated_movies_basics.csv", index=False)
```
